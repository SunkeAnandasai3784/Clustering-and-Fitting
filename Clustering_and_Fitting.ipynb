```python
[X, Y] = np.meshgrid(x, y)

fig, ax = plt.subplots(1, 1)

Z = np.sqrt(X**2+Y**2)

# plots contour lines
ax.contour(X, Y, Z)

ax.set_title('Contour Plot')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')

plt.show()
```
